```python
#!/usr/bin/env python
# coding=utf-8
# -------- import basic package --------
%pylab inline
import pdb
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import warnings
import random
import datetime as dt
import matplotlib.pylab as plt

PATH = 'F:/codeGit/dataset/koubei/dataset/'
FILE_SHOP = 'shop_info.csv'
FILE_PAY = 'df_shopPay.csv'
FILE_VIEW = 'user_view.csv'
```

```python
df_shop = pd.read_csv(PATH + FILE_SHOP)
print df_shop[:10]
```

```python
df_shopPay = pd.read_csv(PATH + FILE_PAY)
print df_shopPay[:10]
print len(df_shopPay)
```

```python
df_shopPay['time'] = pd.to_datetime(df_shopPay['date'], format = '%Y-%m-%d')
```

```python
day_dt = pd.to_datetime('2016.10.31') - pd.to_datetime('2015-07-01')
days = day_dt.days + 1
time_begin = pd.to_datetime('2015-07-01')
x = [(time_begin + dt.timedelta(i)).strftime('%Y-%m-%d') for i in range(days)]
print x

df_temp = pd.DataFrame()
date_std = sort([(time_begin + dt.timedelta(i)).strftime('%Y-%m-%d') for i in range(days)])
for shop in range(2000):
    df_date = pd.DataFrame(sort(date_std), columns = ['date'])
    df_date['shop_id'] = shop + 1 
    df_temp = pd.concat([df_temp, df_date], axis = 0)
    df_temp['temp'] = 0
print df_temp

```

```python
df_shop_fillna = pd.merge(
    df_shopPay[['shop_id', 'date', 'user_pay_cnt']],
    df_temp,
    on = ['shop_id', 'date'], how = 'right'
 ).reset_index()
df_shop_fillna = df_shop_fillna[['shop_id', 'date', 'user_pay_cnt']].fillna(0)
df_shop_fillna['shop_id'] = df_shop_fillna['shop_id'].apply(lambda x : int(x))
print df_shop_fillna
df_shop_fillna.to_csv(PATH + 'df_shop_fillna1.csv', index = False, mode = 'w') 
```

```python
df_shopPay = df_shop_fillna
df_shopPay.columns = ['shop_id', 'time', 'user_pay_cnt']
df_shopPay['time'] = pd.to_datetime(df_shopPay['time'], format = '%Y-%m-%d')
```

```python
train_date_begin = '2016-10-18'
df_shopPay1 = df_shopPay[(df_shopPay.time >= '2016-10-25') & (df_shopPay.time <= '2016-10-31')]
df_shopPay2 = df_shopPay[(df_shopPay.time >= '2016-10-18') & (df_shopPay.time <= '2016-10-24')]
df_shopPay3 = df_shopPay[(df_shopPay.time >= '2016-10-11') & (df_shopPay.time <= '2016-10-17')]
print df_shopPay1.time.unique()
print df_shopPay2.time.unique()
print df_shopPay3.time.unique()
print len(df_shopPay1)
print len(df_shopPay2)
print len(df_shopPay3)
```

```python
results = []
for shop_cur in df_shop['shop_id']:
    df_temp1 = df_shopPay1[df_shopPay1.shop_id == shop_cur]
    df_temp2 = df_shopPay2[df_shopPay2.shop_id == shop_cur]
    df_temp3 = df_shopPay3[df_shopPay3.shop_id == shop_cur]
    
    noise = [random.randint(0,1) for i in range(7)]
    df_temp1.index = df_temp1.time; df_temp2.index = df_temp2.time; df_temp3.index = df_temp3.time
    df_temp1 = df_temp1.sort_index(); df_temp2 = df_temp2.sort_index(); df_temp3 = df_temp3.sort_index()
    result = [shop_cur]

    temp1 = (df_temp1['user_pay_cnt'].values + noise).tolist()
    temp2 = (df_temp2['user_pay_cnt'].values + noise).tolist()
    temp3 = (df_temp3['user_pay_cnt'].values + noise).tolist()
    
    for i in range(7):
        if temp1[i] < 2:
            if temp2[i] > 2 & temp3[i] > 2:
                temp1[i] = int((temp2[i] + temp3[i]) / 2)
            elif temp2[i] <= 2:
                temp1[i] = temp3[i]
            else:
                temp1[i] = temp2[i]
            
    result.extend(temp1)
    meanVal = [x * 1.1 for x in temp1]
    meanVal = meanVal.tolist()
    result.extend(meanVal)
    result = [int(elem) for elem in result]
    result[7] = result[14]
    results.append(result)
pd.DataFrame(results).to_csv(PATH + 'submiss_2017-03-09.csv', index = False, header = False, mode = 'w')
```
